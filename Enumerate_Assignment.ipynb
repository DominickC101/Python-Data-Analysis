```python
location = ['Sun Valley', 'Stowe', 'Mammoth', 'Stowe', 'Sun Valley', 'Mammoth', 'Mammoth', 'Mammoth', 'Sun Valley']
subtotals = [15.98, 899.97, 799.97, 117.96, 5.99, 599.99, 24.99, 1799.94, 99.99]
taxes = []
totals = []

for index, price in enumerate(subtotals):
    if location[index] == 'Sun Valley':
        tax = .08
    elif location[index] == 'Stowe':
        tax = .06
    elif location[index] ==  'Mammoth':
        tax = .0775
    else:
        print('Invalid Location')
    taxes.append(round(price * tax, 2))
    totals.append(round(price + taxes[index], 2))
print(taxes)
print(totals)
```
