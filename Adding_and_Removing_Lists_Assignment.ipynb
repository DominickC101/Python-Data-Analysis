#### Adding Lists Assignment

```python
customer_list = ['C00001', 'C00003', 'C00004', 'C00005', 'C00006', 'C00007', 'C00008', 'C00010', 'C00013', 'C00014', 'C00015', 'C00016', 'C00020']

#Add customer 'C00009' to customer list
customer_list.append('C00009')

#Add saturday_list to updated customer_list
saturday_list = ['C00004', 'C00017', 'C00019', 'C00002', 'C00008', 'C00021', 'C00022']
customer_list += saturday_list
print(customer_list)
```
