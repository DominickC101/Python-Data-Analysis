```python
price_list = [5.99, None, 19.99, 24.99, 0, '74.99', 99.99]

budget = 50

for price in price_list:
    try:
        affordable_quantity = budget // price
        print(f"I can afford {affordable_quantity} of these")
    except ZeroDivisionError:
        print('This item is free')
    except TypeError:
        if price == None:
            continue
        else:
            affordable_quantity = budget // float(price)
            print(f"I can afford {affordable_quantity} of these")
        

        
```

```python
price_list = [5.99, None, 19.99, 24.99, 0, '74.99', 99.99]

budget = 50

for price in price_list:
    if price == None:
        continue
    try:
        affordable_quantity = budget//price
        print(f"I can afford {affordable_quantity} of these")
    except ZeroDivisionError:
        print('This item is free')
    except TypeError:
        affordable_quantity = budget//float(price)
        print(f"I can afford {affordable_quantity} of these")
    except:
        print("That's not a number")
```
